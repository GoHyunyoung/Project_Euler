d 를 1000 이하의 정수라고 할 때, 단위분수 1/d 의 순환마디가 가장 긴 수는 무엇입니까?

```python
def cirPart(n):
    'returns circulation part of 1/n\'s '
    share=list()
    circle=list()
    s=1
    while True:
        share.append(s/n)
        if 0 in circle:
            return 0
        elif s%n in circle:
            return len(circle[circle.index(s%n):len(circle)])
        circle.append(s%n)
        s=(s%n)*10   
```

```python
MAX=tuple([0,0])
for x in range(2,1000+1):
    print x,'의 순환마디의 크기 :',cirPart(x)
    if MAX[1]<cirPart(x):
        MAX=x,cirPart(x)
print 'MAX->',MAX
```
