####위에서 대각선 방향으로 연속된 붉은 숫자 네 개의 곱은 26 × 63 × 78 × 14 = 1788696 입니다.

####그러면 수평, 수직, 또는 대각선 방향으로 연속된 숫자 네 개의 곱 중 최대값은 얼마입니까?

```python
#in:2차 list
#out:2차 list에서 수평으로 곱한 값중의 MAX값
def hrz_MULTImax(Pixel):
    result=0
    for x in range(0,len(Pixel)):
        for y in range(0,len(Pixel[x])-4):
            tmp=1
            for cnt in range(0,4):    #4개의 곱을 tmp로
                tmp*=Pixel[x][y+cnt]
                if result<tmp:
                    result=tmp
            
    return result
```

```python
#in:2차 list
#out:2차 list에서 수직으로 곱한 값중의 MAX값
def vrt_MULTImax(Pixel):
    result=0
    for x in range(0,len(Pixel)-4):
        for y in range(0,len(Pixel[x])):
            tmp=1
            for cnt in range(0,4):    #4개의 곱을 tmp로
                tmp*=Pixel[x+cnt][y]
                if result<tmp:
                    result=tmp
            
    return result
```

```python
#in:2차 list
#out:2차 list에서 오른쪽아래 대각선으로 곱한 값중의 MAX값
def RDdgl_MULmax(Pixel):
    result=0
    for x in range(0,len(Pixel)-4):
        for y in range(0,len(Pixel[x])-4):
            if result<mul([Pixel[x][y],Pixel[x+1][y+1],Pixel[x+2][y+2],Pixel[x+3][y+3]]):
                result=mul([Pixel[x][y],Pixel[x+1][y+1],Pixel[x+2][y+2],Pixel[x+3][y+3]])
    return result
```

```python
#in:2차 list
#out:2차 list에서 왼쪽아래 대각선으로 곱한 값중의 MAX값
def LDdgl_MULmax(Pixel):
    result=0
    for x in range(0,len(Pixel)-4):
        for y in reversed(range(4,len(Pixel[x]))):
            if result<mul([Pixel[x][y],Pixel[x+1][y-1],Pixel[x+2][y-2],Pixel[x+3][y-3]]):
                result=mul([Pixel[x][y],Pixel[x+1][y-1],Pixel[x+2][y-2],Pixel[x+3][y-3]])
    return result
```

```python
max(hrz_MULTImax(intPixel),vrt_MULTImax(intPixel),RDdgl_MULmax(intPixel),LDdgl_MULmax(intPixel))
```
