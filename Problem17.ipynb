1부터 5까지의 숫자를 영어로 쓰면 one, two, three, four, five 이고,

각 단어의 길이를 더하면 3 + 3 + 5 + 4 + 4 = 19 이므로 사용된 글자는 모두 19개입니다.

1부터 1,000까지 영어로 썼을 때는 모두 몇 개의 글자를 사용해야 할까요?

참고: 빈 칸이나 하이픈('-')은 셈에서 제외하며, 단어 사이의 and 는 셈에 넣습니다.

  예를 들어 342를 영어로 쓰면 <i>three hundred and forty-two</i>가 되어서 23 글자,
  
  115 = <i>one hundred and fifteen</i> 의 경우에는 20 글자가 됩니다.
 
 ---

```python
word={1:'one',2:'two',3:'three',4:'four',5:'five',6:'six',7:'seven',8:'eight',9:'nine',
      10:'ten',11:'eleven',12:'twelve',13:'thirteen',14:'fourteen',15:'fifteen',16:'sixteen',17:'seventeen',18:'eighteen',19:'nineteen',
      20:'twenty',30:'thirty',40:'forty',50:'fifty',60:'sixty',70:'seventy',80:'eighty',90:'ninety',100:'hundred',1000:'onethousand'}
```

```python
# int n을 영어로 변환하여 반환
def nEng(n):
    #1의 자리 혹은 10~19 혹은 20,30,40 .. 등 인 경우
    if len(str(n))==1 or len(str(n))==4 or ((len(str(n))==2) and (n/10==1 or n%10==0)) :
        return word[n]
    
    #10의 자리인 경우
    elif len(str(n))==2:
        return word[n/10*10]+'-'+word[n%10]
    
    #100 자리의 경우
    elif len(str(n))==3:
        if n%100 ==0:
            return word[n/100]+'hundred'
        else:
            return word[n/100]+'hundred and '+nEng(n%100)
    return n
```

```python
strCollect=''
for x in range(1,1000+1):
    strCollect+='%s'%(nEng(x))
```

```python
strCollect
```

```python
# ' '여백을 모두 제거
strCollect1=strCollect.replace(' ','')
# '-'문구 모두 제거
strCollect2=strCollect1.replace('-','')
```
